```python
def mc_pitts_neuron(inputs, weights, bias, threshold):
    weighted_sum = sum([x * w for x, w in zip(inputs, weights)]) + bias
    return 1 if weighted_sum >= threshold else 0
```

```python
def and_logic(x1, x2):
    weights = [1, 1]
    bias = 0
    threshold = 2
    return mc_pitts_neuron([x1, x2], weights, bias, threshold)

inputs = [(0, 0), (0, 1), (1, 0), (1, 1)]
print("AND Logic:")
for x1, x2 in inputs:
    print(f"Inputs: {x1}, {x2} -> Output: {and_logic(x1, x2)}")
```

```python
def or_logic(x1, x2):
    weights = [1, 1]
    bias = 0
    threshold = 1
    return mc_pitts_neuron([x1, x2], weights, bias, threshold)

print("OR Logic:")
for x1, x2 in inputs:
    print(f"Inputs: {x1}, {x2} -> Output: {or_logic(x1, x2)}")
```

```python
def not_logic(x):
    weights = [-1]
    bias = 0.5
    threshold = 0
    return mc_pitts_neuron([x], weights, bias, threshold)

print("\nNOT Logic:")
for x in [0, 1]:
    print(f"Input: {x} -> Output: {not_logic(x)}")
```

```python
def nor_logic(x1, x2):
    weights = [-1, -1]
    bias = 0.5
    threshold = 0
    return mc_pitts_neuron([x1, x2], weights, bias, threshold)

print("\nNOR Logic:")
for x1, x2 in inputs:
    print(f"Inputs: {x1}, {x2} -> Output: {nor_logic(x1, x2)}")
```

#### Why Do We Need Multiple Layers?
A single-layer McCulloch-Pitts neuron can only handle linearly separable functions, meaning the input space can be separated by a straight line.
XOR is not linearly separable because the positive examples (0, 1) and (1, 0) cannot be separated from (0, 0) and (1, 1) using a single straight line.
By introducing two layers:

The first layer computes intermediate values (z1 and z2).
The second layer combines these values to produce the XOR output.
This two-layer approach enables us to capture the nonlinear relationship in XOR.

```python
def xor_logic(x1, x2):
    z1 = or_logic(x1, x2)
    z2 = and_logic(x1, x2)
    return and_logic(z1, not_logic(z2))

print("\nXOR Logic:")
for x1, x2 in inputs:
    print(f"Inputs: {x1}, {x2} -> Output: {xor_logic(x1, x2)}")
```
