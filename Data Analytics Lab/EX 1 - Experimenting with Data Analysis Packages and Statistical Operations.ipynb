### Numpy

```python
# 1. Array Creation and Basic Operations

import numpy as np

arr1 = np.array([1, 2, 3, 4])
print("1D Array:", arr1)

arr2 = np.array([[1, 2, 3], [4, 5, 6]])
print("2D Array:\n", arr2)

zeros = np.zeros((2, 3))
print("Array of Zeros:\n", zeros)

ones = np.ones((3, 2))
print("Array of Ones:\n", ones)

range_arr = np.arange(10, 20, 2)
print("Range Array:", range_arr)

linspace_arr = np.linspace(0, 1, 5)
print("Linspace Array:", linspace_arr)
```

```python
sum_arr = arr1 + 5
print("Addition:\n", sum_arr)

# Multiplication
product_arr = arr1 * 2
print("Multiplication:\n", product_arr)

# Element-wise Operations
elementwise_sum = arr2 + arr2
print("Element-wise Sum:\n", elementwise_sum)

# Matrix Multiplication
matrix_product = np.dot(arr2, arr2.T)  # arr2.T is the transpose of arr2
print("Matrix Product:\n", matrix_product)
```

```python
# Mean
mean_val = np.mean(arr1)
print("Mean:", mean_val)

# Median
median_val = np.median(arr1)
print("Median:", median_val)

# Standard Deviation
std_dev_val = np.std(arr1)
print("Standard Deviation:", std_dev_val)

# Variance
variance_val = np.var(arr1)
print("Variance:", variance_val)

# Sum
sum_val = np.sum(arr1)
print("Sum:", sum_val)
```

```python
# Accessing elements in 1D array
print("Element at index 2:", arr1[2])

# Accessing elements in 2D array
print("Element at position (1, 2):", arr2[1, 2])

# Slicing
print("First row of 2D array:", arr2[0])
print("First two rows of 2D array:\n", arr2[:2])
```

```python
# Reshape a 1D array to 2D
reshaped_arr = np.reshape(np.arange(9), (3, 3))
print("Reshaped Array:\n", reshaped_arr)

# Flattening a 2D array to 1D
flattened_arr = reshaped_arr.flatten()
print("Flattened Array:", flattened_arr)
```

```python
# Stacking Arrays Vertically
stacked_v = np.vstack([arr1, arr1])
print("Vertical Stack:\n", stacked_v)

# Stacking Arrays Horizontally
stacked_h = np.hstack([arr1, arr1])
print("Horizontal Stack:", stacked_h)

# Splitting Arrays
split_arr = np.split(np.arange(10), 2)
print("Split Array:", split_arr)
```

### SciPy

```python
from scipy.integrate import quad

# Define the function to integrate
def f(x):
    return np.exp(-x**2)

# Integrate from 0 to infinity
integral, error = quad(f, 0, np.inf)
print("Integral:", integral)
```

```python
from scipy.linalg import solve

# Coefficients matrix
A = np.array([[3, 2], [1, 2]])

# Right-hand side vector
b = np.array([5, 5])

# Solve the system of equations
x = solve(A, b)
print("Solution:", x)
```

```python
from scipy.optimize import minimize

# Define an objective function
def objective(x):
    return x**2 + 5*np.sin(x)

# Initial guess
x0 = 2.0

# Minimize the objective function
result = minimize(objective, x0)
print("Minimum:", result.x)
```

### Statsmodel

```python
#Linear Regression

import statsmodels.api as sm
import numpy as np

X = np.random.rand(100)
y = 2 * X + np.random.randn(100) * 0.1

X = sm.add_constant(X)

model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
# Time Series Analysis
# Example : Autoregressive model.

from statsmodels.tsa.ar_model import AutoReg
import numpy as np

# Generate some time series data
data = np.cumsum(np.random.randn(100))

# Fit an autoregressive model
model = AutoReg(data, lags=1).fit()

# Predict the next value
forecast = model.predict(start=len(data), end=len(data))
print("Forecast:", forecast)
```
